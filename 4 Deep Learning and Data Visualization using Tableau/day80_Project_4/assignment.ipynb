```python
import pandas as pd 
import numpy as np

# Define file path
file_path = "Global Superstore Orders 2016.xlsx"

# Load the Excel file
xls = pd.ExcelFile(file_path)

# Load the "Orders" sheet into a DataFrame
orders_df = pd.read_excel(xls, sheet_name="Orders")

people_df = pd.read_excel(xls, sheet_name="People")


# Set a random seed for reproducibility
np.random.seed(42)

# Assign "Order Returned" as 1 (returned) with a 10% probability, else 0
orders_df["Order Returned"] = np.random.choice([0, 1], size=len(orders_df), p=[0.9, 0.1])
print(orders_df.head())
print("*--------------*")
print(orders_df.info)
print("*--------------*")
print(orders_df.isnull().sum())

print("------------------------")
#print(people_df.head()) 
#print(people_df.info)


# Define the new file path for saving
output_file_path = "Global_Superstore_Orders_Updated.xlsx"

# Save the modified DataFrame to a new Excel file
with pd.ExcelWriter(output_file_path, engine="xlsxwriter") as writer:
    orders_df.to_excel(writer, sheet_name="Orders", index=False)
    people_df.to_excel(writer, sheet_name="People", index=False)

# Provide the file for download
output_file_path

```
