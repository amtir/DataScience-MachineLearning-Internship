```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load a simple dataset (Iris dataset for demonstration)
iris = load_iris()
X = iris.data  # This has 4 features (4D)          Example of the 4 features  Sepal length Sepal width Petal length Petal width
y = iris.target  # Target labels                   There are three species (0, 1, and 2).

# Print original shape of the dataset
print("Original dataset shape:", X.shape)        # Since the Iris dataset has 150 samples (rows) and 4 features (columns), the shape will be (150, 4)

# Let's reduce the dataset to 2 dimensions using PCA
pca = PCA(n_components=2)
# This creates a PCA object that will reduce the dimensionality of the data from 4 features to 2 principal components (or 2D). n_components=2 means we want to reduce the data to 2 dimensions (2 principal components).
X_reduced = pca.fit_transform(X)

# Print the shape after applying PCA
print("Reduced dataset shape:", X_reduced.shape)
# This shows the shape of the dataset after PCA has been applied. Since we reduced the data to 2 dimensions, the shape will be (150, 2).

# Visualize the reduced 2D data
plt.figure(figsize=(8,6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, edgecolor='k', s=100)
# X_reduced[:, 0] and X_reduced[:, 1]: Select the first and second columns of X_reduced (these are the new principal components after PCA).
# c=y: Colors the points according to their target class (iris species).
# edgecolor='k': Adds a black edge around each point.
# s=100: Sets the size of the points to 100.
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Iris Dataset Reduced to 2D')
plt.show()
```
