```python
# Importing necessary libraries
import matplotlib.pyplot as plt  # For data visualization
from sklearn.datasets import make_blobs  # For generating synthetic data points
from sklearn.cluster import KMeans  # For implementing the K-means clustering algorithm

# Step 1: Generate synthetic 2D data with specified clusters
# Here, we're using make_blobs to create 300 data points (n_samples=300)
# 'centers=4' means we want the data to be generated around 4 distinct clusters
# 'cluster_std=0.60' controls the spread (standard deviation) of each cluster
# 'random_state=0' ensures we get the same random data on each run (for reproducibility)
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Step 2: Apply K-means clustering on the generated data
# Define the number of clusters we want to identify as 'k'
k = 4  # Number of clusters

# Initialize the KMeans class with 'n_clusters=k' to define our clusters
# 'random_state=0' is set to make sure the initial centroids are the same on each run
kmeans = KMeans(n_clusters=k, random_state=0)

# Fit the K-means model to our data, and get the cluster labels for each point
# 'fit_predict' finds clusters in X and assigns each point to one of these clusters
y_kmeans = kmeans.fit_predict(X)

# Step 3: Plotting the clustered data on a 2D scatter plot
# Initialize a new figure with a specific size for better readability
plt.figure(figsize=(8, 6))

# Plot each cluster with a different color
# The loop iterates over the number of clusters (range(k))
# For each cluster, we plot points in that cluster (X[y_kmeans == i]) with:
#   - x-coordinates (Feature 1): X[y_kmeans == i, 0]
#   - y-coordinates (Feature 2): X[y_kmeans == i, 1]
# 's=50' sets the size of each point, 'label=f'Cluster {i+1}'' names each cluster for the legend
for i in range(k):
    plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], s=50, label=f'Cluster {i+1}')

# Plotting the centroids of each cluster
# 'kmeans.cluster_centers_' gives the coordinates of the cluster centers
# We plot the x-coordinates (Feature 1) and y-coordinates (Feature 2) of these centroids
# 's=200' sets a larger size for centroids, 'c="black"' makes centroids black for contrast
# 'marker="X"' uses an "X" symbol to visually differentiate centroids from data points
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='black', label='Centroids', marker='X')

# Adding labels and title to the plot for better interpretation
plt.title("K-means Clustering (2D)")  # Title of the plot
plt.xlabel("Feature 1")  # Label for the x-axis
plt.ylabel("Feature 2")  # Label for the y-axis

# Display the legend which shows each cluster and centroid labels
plt.legend()

# Finally, display the plot to visualize the clusters and centroids
plt.show()
```

```python
import numpy as np
import pandas as pd
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0],[5,8],[8,9],[5,7],[6,7]])

wcss =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)


plt.plot(range(1,11), wcss,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
