```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range for the plots
x1 = np.linspace(-10, 10, 1000)
x2 = np.linspace(-10, 10, 1000)

# Avoid division by zero and undefined regions for 1/(x-3)
x1_safe = x1[x1 != 0]
x2_safe = x2[(x2 != 0) & (x2 != 3)]

# Compute y values for the first set of functions
y1 = 1 / x1_safe
y2 = 1 / (x2_safe - 3)

# Define x range for the exponential plots
x_exp = np.linspace(-5, 5, 1000)

# Compute y values for the exponential functions
y_exp = np.exp(x_exp)
y_neg_exp = np.exp(-x_exp)
y_custom_exp = np.exp(-1 / (x_exp - 3))

# Plot 1/x
plt.figure(figsize=(8, 5))
plt.plot(x1_safe, y1, label="1/x")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Graph of 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# Plot 1/(x-3)
plt.figure(figsize=(8, 5))
plt.plot(x2_safe, y2, label="1/(x-3)", color="orange")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Graph of 1/(x-3)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# Plot e^x with proper scaling
plt.figure(figsize=(8, 5))
plt.plot(x_exp, y_exp, label="e^x", color="green")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Graph of e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# Plot e^(-x) with proper scaling
plt.figure(figsize=(8, 5))
plt.plot(x_exp, y_neg_exp, label="e^(-x)", color="red")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Graph of e^(-x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# Plot e^(-1/(x-3)) with proper scaling
plt.figure(figsize=(8, 5))
plt.plot(x_exp[(x_exp - 3) != 0], y_custom_exp[(x_exp - 3) != 0], label="e^(-1/(x-3))", color="purple")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Graph of e^(-1/(x-3))")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.where(x != 3, 5 * np.exp(-1 / (x - 3)) / (x - 3), np.nan)

# Generate x values, avoiding x = 3
x = np.linspace(-30, 30, 1000)
x = x[x != 3]  # Remove the point x = 3

# Calculate y values
y = f(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r"$f(x) = \frac{5e^{-\frac{1}{x-3}}}{x-3}$", color='blue')
plt.axvline(x=3, color='red', linestyle='--', label="Vertical Asymptote $x=3$")
plt.axhline(y=0, color='black', linestyle='--', label="Horizontal Asymptote $y=0$")
plt.title("Graph of $f(x) = \\frac{5e^{-\\frac{1}{x-3}}}{x-3}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(-10, 10)  # Adjust y-limits for better visualization
plt.grid()
plt.legend()
plt.show()
```
