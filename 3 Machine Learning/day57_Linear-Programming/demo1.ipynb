```python
# Gradient descent example
import numpy as np

# Function and its gradient
def f(x):
    return x**2 + 4*x + 4

def grad_f(x):
    return 2*x + 4

# Gradient descent
x = 0  # Initial guess
eta = 0.1  # Learning rate
for _ in range(100):
    x = x - eta * grad_f(x)
    print(f"x: {x}, f(x): {f(x)}")
```

```python
# Re-import necessary libraries after reset
import numpy as np
import matplotlib.pyplot as plt

# Define the function, its first derivative, and second derivative
def f(x):
    return x**3 - 3*x**2 + 2

def f_prime(x):
    return 3*x**2 - 6*x

def f_double_prime(x):
    return 6*x - 6

# Generate x values
x = np.linspace(-1, 3, 500)
y = f(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = x^3 - 3x^2 + 2", linewidth=2)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")

# Critical points (where f'(x) = 0)
critical_points = [0, 2]
for cp in critical_points:
    plt.scatter(cp, f(cp), color="red", label=f"Critical Point (x={cp}, f(x)={f(cp)})")

# Annotate max and min based on second derivative
plt.text(-0.5, f(0) + 1, "Maximum", fontsize=12, color="red")
plt.text(2.1, f(2) - 1.5, "Minimum", fontsize=12, color="blue")

plt.title("First vs. Second Derivatives (Critical Points)", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
