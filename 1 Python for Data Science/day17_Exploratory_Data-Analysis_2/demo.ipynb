```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data from the uploaded csv file
data = pd.read_csv("cereal.csv", sep = ',')
# 1 Explore data
data.info()
data.describe()
```

```python
# Calculate the top five cereal manufacturers based on average customer rating
top_five_manufacturers = data.groupby("mfr")["rating"].mean().sort_values(ascending=False).head(5)
#top_five_manufacturers
# Display the result to the user
print("Top Five Cereal Manufacturers by Rating", top_five_manufacturers)
```

```python
# Define the dictionary with manufacturer names
manufacturer_names = {
    'A': 'American Home Food Products',
    'G': 'General Mills',
    'K': 'Kelloggs',
    'N': 'Nabisco',
    'P': 'Post',
    'Q': 'Quaker Oats',
    'R': 'Ralston Purina'
}

# Replace the manufacturer codes with full names in the data
data['mfr'] = data['mfr'].map(manufacturer_names)

# Display the updated data with replaced manufacturer names
data.head()
```

```python

# Sort data by high protein and low calories
sorted_data = data.sort_values(by=['protein', 'calories'], ascending=[False, True])

# Create scatter plot to visualize high-protein, low-calorie cereals
plt.figure(figsize=(10, 6))
plt.scatter(sorted_data['calories'], sorted_data['protein'], alpha=0.7, c='blue')

# Label manufacturers on points where protein is high and calories are low
for i, row in sorted_data.iterrows():
    if row['protein'] >= sorted_data['protein'].quantile(0.75) and row['calories'] <= sorted_data['calories'].quantile(0.25):
        plt.text(row['calories'], row['protein'], row['mfr'], fontsize=8, ha='right')

# Set plot labels and title
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.title('Cereal Manufacturers with High Protein and Low Calories')
plt.grid(True)
plt.show()

print("Top Cereals", top_cereals)
```

```python
# Step 1: Select the top N cereals with high protein and low calories (change `n` to 10 for top 10)
n = 10
top_cereals = data.sort_values(by=['protein', 'calories'], ascending=[False, True]).head(n)

# Step 2: Define a color map
colors = plt.get_cmap('tab10', n)  # 'tab10' is a colormap with 10 distinct colors

# Step 3: Plot each cereal with a color from the color map and add to legend
plt.figure(figsize=(10, 6))
for i, (index, row) in enumerate(top_cereals.iterrows()):
    plt.scatter(row['calories'], row['protein'], color=colors(i), label=f"{row['name']} ({row['mfr']})", s=100)

# Set plot labels and title
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.title(f'Top {n} Cereal Products with High Protein and Low Calories')

# Position the legend outside the plot to the right
plt.legend(title="Cereal (Manufacturer)", loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()

print(f"Top {n} Cereals", top_cereals)
```

```python

# Selecting only nutritional columns for correlation and covariance analysis
nutritional_columns = ['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']

# Compute correlation and covariance matrices
correlation_matrix = data[nutritional_columns].corr()
covariance_matrix = data[nutritional_columns].cov()

# Plotting the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={'label': 'Correlation Coefficient'})
plt.title("Correlation Heatmap of Nutritional Values")
plt.show()

# Plotting the covariance heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(covariance_matrix, annot=True, cmap="YlGnBu", fmt=".2f", cbar_kws={'label': 'Covariance Value'})
plt.title("Covariance Heatmap of Nutritional Values")
plt.show()
```
