```python
import pandas as pd

# Example data
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'Purchased': ['Yes', 'No', 'Yes', 'Yes', 'No', 'No']
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# Perform cross-tabulation
cross_tab = pd.crosstab(df['Gender'], df['Purchased'])
print(cross_tab)
```

```python
#Covariance
import pandas as pd
x = [100, 200, 137, 457, 890, 345]
y = [345, 678, 123, 56, 890, 500,120,230]

X = pd.Series(x)
Y = pd.Series(y)
print(X.cov(Y))
```

```python
#Correlation
print(X.corr(Y))
```

```python
#Correlation matrix
x = [100, 200, 137, 457, 890, 345]
y = [345, 678, 123, 56, 890, 500]
data = pd.DataFrame({
    'x': x,
    'y': y
})
correlation_matrix = data.corr()

print("Correlation Matrix:")
print(correlation_matrix)

'''Diagonal Values (1): Each variable is perfectly correlated with itself.
   Off-Diagonal Value (0.553): Represents the Pearson correlation coefficient between x and y.'''
```

```python
# Create a heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(4,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0,
            xticklabels=['x', 'y'], yticklabels=['x', 'y'])
plt.title('Correlation Matrix Heatmap')
plt.show()

'''annot=True adds numerical annotations to each cell.
   cmap specifies the color map ('viridis' or 'coolwarm').
   linewidths and linecolor control the appearance of cell borders.
   vmin and vmax set the color scale limits.
   center centers the color map around zero, useful for correlation matrices.'''
```

```python
# Create a scatter plot
plt.figure(figsize=(4,4))
plt.scatter(x, y, color='blue', marker='o', edgecolor='k')

# Add titles and labels
plt.title('Scatter Plot of x vs y')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Define the data for perfect positive correlation
import numpy as np
x = np.array([1, 2, 3, 4, 5, 6])
y = 2 * x + 1  # y is perfectly correlated with x

# Create a scatter plot
plt.figure(figsize=(4,4))
plt.scatter(x, y, color='blue', marker='o', edgecolor='b')

# Add the line of best fit
plt.plot(x, y, color='red', linestyle='--')

# Add titles and labels
plt.title('Perfect Positive Linear Correlation')
plt.xlabel('x')
plt.ylabel('y')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Define the data for perfect negative correlation
x = np.array([1, 2, 3, 4, 5, 6])
y = -2 * x + 13  # y is perfectly negatively correlated with x

# Create a scatter plot
plt.figure(figsize=(4,4))
plt.scatter(x, y, color='green', marker='o', edgecolor='b')

# Add the line of best fit
plt.plot(x, y, color='orange', linestyle='--')

# Add titles and labels
plt.title('Perfect Negative Linear Correlation')
plt.xlabel('x')
plt.ylabel('y')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Define the data for strong positive correlation
'''the data will have a strong positive linear relationship, where the correlation coefficient is close to  +1'''

np.random.seed(0)  # For reproducibility
x = np.linspace(1, 10, 100)
y = 2 * x + np.random.normal(0, 1, size=x.shape)  # Strong positive correlation with some noise
#y is generated as 2x plus some normally distributed noise, resulting in a strong positive relationship.

# Create a scatter plot
plt.figure(figsize=(4,4))
plt.scatter(x, y, color='blue', marker='o', edgecolor='b')

# Add the line of best fit
m, b = np.polyfit(x, y, 1)  # Linear fit
plt.plot(x, m*x + b, color='red', linestyle='--')

# Add titles and labels
plt.title('Strong Positive Linear Correlation')
plt.xlabel('x')
plt.ylabel('y')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(5, 2, 100),
    'C': np.random.normal(-3, 0.5, 100),
    'D': np.random.normal(10, 3, 100)
})

# Create the scatter plot matrix
sns.pairplot(data) 
'''generates the scatter plot matrix. Each cell in the matrix is a scatter plot showing the relationship between a pair of variables. 
The diagonal usually contains histograms or density plots of individual variables.'''

# Add a title
plt.suptitle('Scatter Plot Matrix', y=1.02)  # y=1.02 to move the title up

# Display the plot
plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(4,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='b', vmin=-1, vmax=1, center=0)

# Add titles and labels
plt.title('Correlation Matrix Heatmap')
plt.xlabel('Variable')
plt.ylabel('Variable')

# Display the plot
plt.show()
```
