```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(0)

#normal distribution

mean =50
std = 10
size = 1000

data = np.random.normal(mean,std,size)

#plot

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)

plt.hist(data,bins=30,density=True,alpha = 0.6, color = 'skyblue',edgecolor='black')

plt.title('Normal')

plt.xlabel('Value')

plt.ylabel('Frequency')

 

#overlaying the theoretical normal distribution curve

xmin , xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = stats.norm.pdf(x,mean,std)
plt.plot(x,p, 'k', linewidth=2)
plt.show()

#2 z distribution

zscores = (data-mean)/std

plt.subplot(1,2,2)
plt.hist(zscores,bins=30,density=True,alpha = 0.6, color = 'salmon',edgecolor='black') 
plt.title('Z- Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

#overlaying the standard normal distribution curve

xmin , xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = stats.norm.pdf(x,0,1)
plt.plot(x,p, 'k', linewidth=2)

plt.show()
```
