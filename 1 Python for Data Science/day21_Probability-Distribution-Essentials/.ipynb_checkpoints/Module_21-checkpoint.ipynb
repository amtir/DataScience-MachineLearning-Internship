```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a non-normal population
np.random.seed(42)  # For reproducibility
population_size = 10000
population = np.random.exponential(scale=2.0, size=population_size)  # Exponential distribution (not normal)

# Step 2: Draw samples and calculate sample means
sample_size = 30  # Sample size per sample
num_samples = 1000  # Number of samples to draw
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Step 3: Calculate standard error
population_std = np.std(population)
standard_error = population_std / np.sqrt(sample_size)

# Step 4: Plot population distribution and sampling distribution of sample means
plt.figure(figsize=(14, 6))

# Population distribution (Exponential)
plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color='skyblue')
plt.title("Population Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Sampling distribution of sample means (CLT in action)
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='salmon')
plt.title(f"Sampling Distribution of Sample Means\n(Sample Size = {sample_size}, Num Samples = {num_samples})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Display calculated standard error
standard_error
```
