```python
#Q3. Problem Statement: Measures of Central Tendency- III
books = pd.read_csv("Books.csv", index_col=False)
books.info()
books.head()
```

```python
# Step 1: Perform initial analysis
books.describe()
```

```python
# Step 2: Group by 'Author' and calculate average 'User Rating' and 'Price'
author_grouped = books.groupby('Author').agg(
    avg_user_rating=('User Rating', 'mean'),
    avg_price=('Price', 'mean'),
    total_books=('Name', 'count')  # Also calculate the total number of books per author
).reset_index()

print(author_grouped.head())
```

```python
# Step 3: Find the author with the maximum number of books
most_books_author = author_grouped.loc[author_grouped['total_books'].idxmax()]
print("Author with maximum books:", most_books_author)
```

```python
# Sort by 'avg_price' in descending order (highest price first)
author_grouped_sorted = author_grouped.sort_values(by='avg_price', ascending=False)

# Display the sorted data
print(author_grouped_sorted.head())
```

```python
# Sort by 'total_books' in descending order (highest total_books first)
author_grouped_sorted = author_grouped.sort_values(by='total_books', ascending=False)

# Display the sorted data
print(author_grouped_sorted.head())
```

```python
# Sort by 'avg_user_rating  ' in descending order
author_grouped_sorted = author_grouped.sort_values(by='avg_user_rating', ascending=False)

# Display the sorted data
print(author_grouped_sorted.head(25))
```
