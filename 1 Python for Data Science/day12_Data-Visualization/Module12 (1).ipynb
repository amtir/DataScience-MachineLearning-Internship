```python
#Grouped bar chart using width parameter

# Data
products = ['Product A', 'Product B', 'Product C']
sales_2023 = [300, 250, 400]
sales_2024 = [320, 270, 380]
n = len(products)
bar_width = 0.35
index = np.arange(n) # Positions of the bars
fig, axis = plt.subplots() #The figure is the overall container for all the plot elements. It holds the subplots (axes), titles, and other elements.

# Bars for 2023
bars_2023 = axis.bar(index - bar_width/2, sales_2023, bar_width, label='2023', color='b')

# Bars for 2024
bars_2024 = axis.bar(index + bar_width/2, sales_2024, bar_width, label='2024', color='r')

axis.set_xlabel('Products')
axis.set_ylabel('Sales')
axis.set_title('Sales Comparison by Product and Year')
axis.legend()
```

```python
#HISTOGRAM

import pandas as pd
data = {
    'Student Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [85, 78, 92, 88, 95],
    'Science': [90, 82, 85, 80, 88],
    'English': [88, 74, 91, 84, 92],
    'History': [76, 80, 85, 89, 94]
}
df = pd.DataFrame(data)
# Plot histograms for each subject
subjects = ['Math', 'Science', 'English', 'History']

plt.figure(figsize=(12, 8)) # used to create a new figure with a specified size.
                            # 12 is the width of the figure in inches.
                            # 8 is the height of the figure in inches.

for i, subject in enumerate(subjects):
    plt.subplot(2, 2, i+1)  # Create a 2x2 grid of subplots
    plt.hist(df[subject], bins=5, edgecolor='black', alpha=0.7) #alpha- used to set the transparency level of plot elements.
                                                                #(0 is fully transparent (completely invisible). and (1 is fully opaque (completely visible).))
    plt.title(f'{subject} Marks Distribution')
    plt.xlabel('Marks')
    plt.ylabel('Frequency')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
#PIE CHART

labels = ['Company A', 'Company B', 'Company C', 'Company D']
sizes = [30, 25, 35, 10]  # Market share percentages
colors = ['gold', 'lightskyblue', 'lightcoral', 'lightgreen']  # Colors for each slice
plt.figure(figsize=(4, 4))  # Set figure size
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=0)
plt.title('Market Share Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Define which slices to "explode" (highlight)
explode = (0.1, 0, 0, 0)  # Explode the first slice
plt.pie(sizes, labels=labels, rotatelabels=True, colors=colors, autopct='%1.1f%%', startangle=0,explode=explode)
plt.title('Market Share Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#SCATTER PLOT

x = [10, 20, 30, 40, 50]  # X-axis values
y = [15, 25, 35, 45, 55]  # Y-axis values
sizes = [50, 100, 150, 200, 550]  # Sizes of scatter points
colors = ['red', 'blue', 'green', 'purple', 'orange']  # Colors of scatter points

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, edgecolors='w')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot Example')
```

```python
#BOX PLOT

data = [45, 67, 56, 78, 90, 45, 67, 89, 92, 55, 77, 88, 99, 74, 63, 150, 200]

# Create the horizontal boxplot
plt.figure(figsize=(10, 6))  # Set figure size
plt.boxplot(data, vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'),
            flierprops=dict(marker='o', color='green', alpha=0.7))

'''patch_artist=True: Enables coloring of the boxes.
boxprops=dict(facecolor='lightblue', color='blue'): Sets the box color to light blue and the border color to blue.
medianprops=dict(color='red'): Colors the median line in red.
vert=False : Sets the orientation of the boxplot to horizontal.'''

# Add labels and title
plt.xlabel('Scores')
plt.title('Horizontal Box Plot of Exam Scores with Outliers')

# Display the boxplot
plt.show()
```

```python

df = sns.load_dataset('iris')

print(df)

fig,axis = plt.subplots(2,2,figsize=(8,8))

plt.subplots_adjust(hspace=0.5,wspace=0.5)

sns.scatterplot(data=df,x='sepal_length',y='sepal_width',hue='species',ax=axis[0,0])

axis[0,0].set_title('Scatter Plot')

sns.boxplot(data=df,x='species',y='petal_length',ax=axis[0,1])

axis[0,1].set_title('Box Plot')

sns.histplot(data=df,x='petal_width',hue='species',ax=axis[1,0])

axis[1,0].set_title('Histogram')

sns.barplot(data=df,x='species',y='petal_length',ax=axis[1,1])

axis[1,1].set_title('Bar Plot')

plt.show()
```

```python
# Import the necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the 'iris' dataset from Seaborn
iris = sns.load_dataset('iris')

# Drop the non-numerical 'species' column
iris_numeric = iris.drop(columns=['species'])

# Compute the correlation matrix on numerical data
correlation_matrix = iris_numeric.corr()

# Set the size of the plot
plt.figure(figsize=(8, 6))

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Show the plot
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```
