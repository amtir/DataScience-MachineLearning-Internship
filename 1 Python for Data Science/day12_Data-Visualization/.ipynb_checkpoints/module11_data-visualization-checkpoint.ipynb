```python
import matplotlib.pyplot as plt 
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
categories=['A','B','C','D']
values=[10,20,30,40]
plt.bar(categories , values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
categories=['A','B','C','D']
values=[10,20,30,40]
plt.barh(categories , values)
plt.xlabel('Values ')
plt.ylabel('Categories')
plt.show()
```
