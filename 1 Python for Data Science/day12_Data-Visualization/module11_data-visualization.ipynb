```python
import matplotlib.pyplot as plt 
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
categories=['A','B','C','D']
values=[10,20,30,40]
plt.bar(categories , values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
categories=['A','B','C','D']
values=[10,20,30,40]
plt.barh(categories , values)
plt.xlabel('Values ')
plt.ylabel('Categories')
plt.show()
```

```python
# Creating the scatter plot instead of a line plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Plot scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Corrected code
size = [15, 30, 45, 10]
lbl = ['A', 'B', 'C', 'D']

# Plot pie chart
plt.pie(size, labels=lbl, autopct='%1.1f%%')  # Fix: changed '%1.1f%' to '%1.1f%%' for correct formatting
plt.title('Pie Chart')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("SampleSuperstore.csv")
print(df)
```

```python
#LINE PLOT

plt.plot(df.Profit)
```

```python
plt.plot(df.Profit[-20:]) #It selects the last 20 rows of the Profit column.
```

```python
#With List

# Number of hours studied
hours = [1, 2, 3, 4, 5]

# Corresponding test scores
scores = [50, 55, 60, 65, 70]

# Create the line plot
plt.plot(hours, scores, marker='o')  # 'marker='o'' adds a circle at each data point

# Add labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Test Score vs. Hours Studied')
```

```python
#With dictionary

# Sales data
sales_data = {'January': 150,'February': 200,'March': 180,'April': 220,'May': 250,
              'June': 270,'July': 300,'August': 280,'September': 230,'October': 210,'November': 190,'December': 240}

# Extract months and sales figures
months = list(sales_data.keys())
sales = list(sales_data.values())

# Create the line plot
plt.plot(months, sales, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data')
plt.xticks(rotation=45)  # Rotate month labels for better readability
```

```python
#BAR CHART

data = df[0:11]
plt.bar(data.index,data.Sales,label='Sales')
plt.bar(data.index,data.Profit,label='Profit')
plt.legend()
```

```python
#Horizontal Bar Chart

plt.barh(data.index,data.Sales,label='Sales')
plt.barh(data.index,data.Profit,label='Profit')
plt.legend()
```
